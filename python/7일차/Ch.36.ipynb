raise 구문
    * 예외를 강제로 발생시키는 기능, 우리가 의도하지 않게 작동하는 상황을 방지하기 위해 사용
사용 방법 : raise 뒤에 예외 이름을 이벽
raise 예외 객체

```python
number = input("숫자를 입력해주세요")
number = int(number)

if number > 0:
    raise NotImplementedError #아직 미구현 상태다
else:
    raise NotImplementedError
```

# datetime 모듈

```python
import datetime

print("오늘 날짜와 시간 출력")
now = datetime.datetime.now()
print(now.year, "년")
print(now.month, "월")
print(now.day, "일")
print(now.hour, "시")
print(now.minute, "분")
print(now.second, "초")

```
